Q-Learning

```python
Q=np.zeros([n_state,n_action])
```

```python
Q
```

```python
episodes=1
G=0          # goal state
alpha=0.618  # it is changeble
```

```python

for episode in range(1, episodes + 1):
    state, info = env.reset()
    done = False
    G = 0
    print("Initial state = {}".format(state))

    while not done:
        action = np.argmax(Q[state])  # greedy action
        next_state, reward, terminated, truncated, info = env.step(action)

        # Q-learning update
        Q[state, action] += alpha * (reward + np.max(Q[next_state]) - Q[state, action])

        G += reward
        state = next_state
        done = terminated or truncated  # episode ends

    print("Episode {} total reward {}".format(episode, G))
```
